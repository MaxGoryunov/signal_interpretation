```python
import random


q = 25
mod = 24

divisors = [i for i in range(1, q) if mod % i == 0]
buckets = {div: [] for div in divisors}

for e in range(1, q):
    print(f'X^s = X^{e}')
    msg = ''
    for div in divisors:
        res = e * div % mod
        if res == 0:
            msg += f'X^({e}*{div}) = X^{res}'
            print(msg)
            buckets[div].append(e)
            break
        else:
            msg += f'X^({e}*{div}) = X^{res}; '

print('Distribution:')
print(buckets)
prim_e = buckets[mod][random.randint(0, len(buckets[mod]))]
print(f'Primitive element: {prim_e = }')
cum = 1
for i in range(1, q):
    cum = (cum * prim_e) % mod 
    print(f'{prim_e}^{i} = {cum}')
```
